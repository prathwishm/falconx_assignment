Read the csv file using pandas

```python
import pandas as pd
df = pd.read_csv('BTCUSDT-trades-2022-03.csv')
df.head()
```

Looks like column names are not specified. Lets specify the column names and load the data again

```python
df = pd.read_csv('BTCUSDT-trades-2022-03.csv', header=None, names = ['trade Id',	'price', 'qty', 'quoteQty','time', 'isBuyerMaker','isBestMatch'])
df.head()
```

The time is present as a timestamp, lets convert it to Datetime and store it in a new column. (this is required to filter only the first 3 days of March and backtrader also needs a datetime column). The time stamp is in utc time zone.

```python
import datetime
df['dt'] = df['time'].apply(lambda x: datetime.datetime.utcfromtimestamp(x/1000))
df.head()
```

Lets Filter the data for the first 3 days of March 2022.

```python
df = df[(df['dt'] >= '2022-03-01') & (df['dt'] < '2022-03-4')]
df
```

Lets set the 'dt' column as the index

```python
df.set_index('dt', inplace=True)
df.head()
```

Since only Price and time is required for backtesting the strategy, lets remove all the unnecessary columns

```python
df = df['price'].to_frame()
df.head()
```

Lets save the dataframe to a csv file for further use

```python
df.to_csv('BTCUSDT-trades-2022-03_price.csv')
```
